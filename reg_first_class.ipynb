```python
x=np.arange(1,21)
```

```python
x   #thhis is our main x axis data 
```

```python
y=x**3
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#for the 1 polynomial degree of the funtion...

w0=.3
w1=.5
k=0
lr=.001
while(k<50):
    k=k+1  
    yp=w0+w1*x
    loss=1/len(x)*sum((yp-y)**2)
    print(loss)
    w0=w0-lr*(2/len(x)*sum(yp-y))
    w1=w1-lr*(2/len(x)*sum((yp-y)*x))
    plt.scatter(x,y,color="red")
    plt.plot(x,yp)
    
```

```python
#this is the for the 2 polynomial degree ...

w0=.3
w1=.5
w2=.6
k=0
lr=.00001
while(k<50):
    k=k+1  
    yp=w0+w1*x+w2*x**2
    loss=1/len(x)*sum((yp-y)**2)
    print(loss)
    w0=w0-lr*(2/len(x)*sum(yp-y))
    w1=w1-lr*(2/len(x)*sum((yp-y)*x))
    w2=w2-lr*(2/len(x)*sum((yp-y)*x**2))
    plt.scatter(x,y,color="red")
    plt.plot(x,yp)
    
```

```python
w0=.3
w1=.5
w2=.6
w3=.7
w4=.8
k=0
lr=.0000000001
while(k<50):
    k=k+1  
    yp=w0+w1*x+w2*x**2+w3*x**3+w4*x**4
    loss=1/len(x)*sum((yp-y)**2)
    print(loss)
    w0=w0-lr*(2/len(x)*sum(yp-y))
    w1=w1-lr*(2/len(x)*sum((yp-y)*x))
    w2=w2-lr*(2/len(x)*sum((yp-y)*x**2))
    w3=w3-lr*(2/len(x)*sum((yp-y)*x**3))
    w4=w4-lr*(2/len(x)*sum((yp-y)*x**4))
    plt.scatter(x,y,color="red")
    plt.plot(x,yp)
    
```
